```python
eraseName = ['Max1','Max2','Max3','Max4','Max5','Min1','Min2','Min3','Min4','Min5']
```

```python
data = data.drop(eraseName,axis=1)
```

## Data Analysis 

```python
copyData = data.copy()
x_data_list = ['Power(W)','Speed']
dummy_erase_list = ['Keyhole induced pore']
```

```python
x_data = copyData[x_data_list]
y_data = copyData.drop(dummy_erase_list, axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data, test_size=0.2)
```

```python
print(x_train)
```

```python
num_x_input = x_train.shape[1]
num_y_input = y_train.shape[1]
print(num_x_input)
print(num_y_input)
```

```python
optimizer_function = keras.optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)
```

```python
model = Sequential()
model.add(Dense(4, input_dim=2,activation='relu'))
model.add(Dense(6, activation='relu'))
#model.add(Dense(6, activation='sigmoid'))
#model.add(Dense(3, activation='relu'))
model.add(Dense(8))
```

```python
model.compile(optimizer=optimizer_function, loss='mse', metrics=["accuracy"])
hist = model.fit(x_train, y_train, epochs=100000)
```

```python
import matplotlib.pyplot as plt
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()
loss_ax.set_ylim([0.0, 1.0])
acc_ax.set_ylim([0.0, 1.0])

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
acc_ax.plot(hist.history['acc'], 'b', label='train acc')
loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')
loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

n = 0
print("문제 : ",x_test[n].reshape(1,4))
print(model.predict(x_test[n].reshape(1,4)))
```
